```python

import torch
from ultralytics import YOLO


def load_yolo_model(config_path):
    
    """
     loading the yolo model intially with pretrained weights

    Returns:
        YOLO model
    """
    return YOLO(config_path)

def train_model_with_yolo(model, data_config, num_epochs):
    model.train(data=data_config, epochs=num_epochs)

def main():
    
    """
    Run the YOLO model dor specified list classes in data.yaml file
    """
    yolo_config = "YOLOv8n.yaml"
    data_config = "data.yaml"
    num_epochs = 1000

    if torch.cuda.is_available():
        yolo_model = load_yolo_model(yolo_config)
        train_model_with_yolo(yolo_model, data_config, num_epochs)
    else:
        print("CUDA is not available on this device.")

if __name__ == "__main__":
    main()
    
```
